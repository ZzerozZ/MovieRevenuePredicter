```python
from chrome_crawler import Chrome_Crawler
from lib import *
```

```python
STUDIO = 'https://www.boxofficemojo.com/studio/chart/?studio=%s.htm'
```

```python
STUDIO_XPATH = {
    'name': '/html[1]/body[1]/div[1]/div[3]/div[2]/p[1]/b[1]/font[1]',
    'totals': '/html[1]/body[1]/div[1]/div[3]/div[2]/table[3]/tbody[1]/tr[%d]/td[2]/font[1]',
    'averages': '/html[1]/body[1]/div[1]/div[3]/div[2]/table[3]/tbody[1]/tr[%d]/td[2]/font[1]',
    'numbers': '/html[1]/body[1]/div[1]/div[3]/div[2]/center[1]/font[1]/b[1]/font[1]'
    }
```

```python
_chrome_option = webdriver.ChromeOptions()
_chrome_option.add_argument("headless")
chrome_option = _chrome_option
driver = webdriver.Chrome(chrome_options=chrome_option)
```

```python
def get_list_studio():
    listPersons = os.listdir('Persons')
    listStudio = []
    
    for person in listPersons:
        # Verify person is .json file
        if (not(person.endswith('.json'))):
            break
            
        # Create link of person
        url = 'Persons\\%s' %person
        # Get person information
        personInfo = json.load(open(url, 'r'))
        # Get list movies from person
        movies = personInfo['movies']
        for movie in movies:
            listStudio.append(movie['studio']) 
        

    # Remove duplicated values
    studios = np.unique(listStudio)
    return studios
```

```python
studios = get_list_studio()
studios = np.delete(studios,0)
studios
```

```python
def create_url(studio):
    return STUDIO % studio
```

```python
def get_studio_info(url):
    driver.get(url)
    
    id = url.split('studio=')[1][:-4]
    
    
    try:
        name = driver.find_element_by_xpath(STUDIO_XPATH['name']).text
        numbers = driver.find_element_by_xpath(STUDIO_XPATH['numbers']).text[3:]
        numbers = int(numbers)
        totals = driver.find_element_by_xpath(STUDIO_XPATH['totals'] %(numbers+3)).text.split('$')[1]
        averages = driver.find_element_by_xpath(STUDIO_XPATH['averages'] %(numbers+4)).text.split('$')[1]
#     totals = ''
#     averages = ''
    except:
       return {
           'id':id,
           'name':'',
           'totals':'',
           'averages':''
       }
        
    studio = {
        'id':id,
        'name':name,
        'totals':totals,
        'averages':averages
    }
    
    return studio
    
```

```python
_round = 0
chrome_crawler = Chrome_Crawler()

while _round < len(studios):
    studioInfo = get_studio_info(create_url(studios[_round]))
    id = studioInfo['id']
    if (studioInfo['totals'] == '' or studioInfo['averages'] == ''):
        print('exception: %s' % id)
        with open('StudioExceptions.txt', 'a+') as f_except:
            f_except.write(id + '\n')
    else:
        try:
            with open('Studios/%s.json' % id, 'w') as fp:
                json.dump(studioInfo, fp)

            time.sleep(np.random.randint(1,4)/4)

            # Reset driver:
            if _round % 10 == 0:
                print ("Crawling studio...\t%d/%d\n"%(_round, len(studios)))
                chrome_crawler.restart_crawler(delay_time=15)
        except:
            print('exception: %s' % id)
            with open('StudioExceptions.txt', 'a+') as f_except:
                f_except.write(id + '\n')
            
    _round += 1
```
