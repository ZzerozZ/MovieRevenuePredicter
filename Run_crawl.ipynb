```python
movies = load_obj('/home/nghia/Desktop/Lab/Movie/obj/all_movie')
movies = [movies[i] for i in range(len(movies)) if i not in [3, 160]]
movie_list = []
studio_list = []
for i in range(len(movies)):
    movie_list += movies[i][1]
    studio_list += movies[i][-1]
```

```python
df = pd.DataFrame([movie_list, studio_list]).T
df.columns = ['movie', 'studio']
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
save_obj(df, 'obj/df_movie_studio')
```

```python
movie_crawler = Movie_Crawler()

_round = 1
for movie in df.movie.values:
    try:
        movie_info = movie_crawler.get_movie_info(movie, num_actor=5)
        with open('Movies/%s.json'%movie, 'w') as fp:
            json.dump(movie_info, fp)

        time.sleep(np.random.randint(1,4)/4)

        # Reset driver:
        if _round % 50 == 0:
            print ("Crawling movie...\t%d/%d\n"%(_round, df.movie.shape[0]))
            movie_crawler.restart_crawler(delay_time=15)
        _round += 1
    except:
        with open('exceptions.txt', 'a+') as f_except:
            f_except.write(movie + '\n')
```
